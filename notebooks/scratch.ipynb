```python
import pandas as pd
import json

# Load the dataset
file_path = '../data/combined_df.csv'
data = pd.read_csv(file_path, low_memory=False)

# Select the necessary features
features = ['total_mass', 'span', 'period_mins', 'perigee_km', 'apogee_km',
            'inclination', 'object_type', 'status', 'object_id']

# Extract a random sample of 10 rows
sample = data[features].sample(n=10)

# Convert the sample to a list of dictionaries
sample_dicts = sample.to_dict(orient='records')

# Convert the list of dictionaries to JSON format
sample_json = json.dumps(sample_dicts, indent=4)

# Print the JSON
print(sample_json)
```

```python
import os

# Set the size limit (100 MB)
size_limit = 100 * 1024 * 1024

# Get the current directory
current_dir = os.getcwd()

# Walk through the directory
for root, dirs, files in os.walk('..'):
    for file in files:
        file_path = os.path.join(root, file)
        file_size = os.path.getsize(file_path)
        if file_size > size_limit:
            print(f"{file_path} - {file_size / (1024 * 1024):.2f} MB")
```
