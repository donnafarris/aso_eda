```python
import pandas as pd
import json

# Load the dataset
file_path = '../data/combined_df.csv'
data = pd.read_csv(file_path, low_memory=False)

# Select the necessary features
features = ['total_mass', 'span', 'period_mins', 'perigee_km', 'apogee_km',
            'inclination', 'object_type', 'status', 'object_id']

# Filter data based on a specific status value
specific_status = 'L'  # Change this to the desired status value
filtered_data = data[data['status'] == specific_status]

# Extract a random sample of 10 rows from the filtered data
sample = filtered_data[features].sample(n=10)

# Convert the sample to a list of dictionaries
sample_dicts = sample.to_dict(orient='records')

# Convert the list of dictionaries to JSON format
sample_json = json.dumps(sample_dicts, indent=4)

# Print the JSON
print(sample_json)
```
